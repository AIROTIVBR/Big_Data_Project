```python
df = pd.read_csv("df_states2020.csv")
df
```

```python
colunas_excluir =['codmun','codRegiaoSaude','nomeRegiaoSaude','casosAcumulado','casosNovos','Recuperadosnovos','emAcompanhamentoNovos','semanaEpi','Unnamed: 0','municipio','coduf','populacaoTCU2019']
df=df.drop(colunas_excluir, axis=1)
```

```python
df
```

```python
nordeste = df[df.regiao == 'Nordeste']
nordeste
```

```python
nordeste.data = pd.to_datetime(nordeste.data)
```

```python
nordeste['data_mes'] = nordeste.data.dt.to_period('M').astype(str)
```

```python
mortes_mes = nordeste.groupby(by='data_mes').obitosNovos.sum()
```

```python
nordeste = [go.Bar(x=mortes_mes.index,
               y=mortes_mes.values,
               marker= {'color': 'lavender'})]

#criando layout
configuracoes_layout = go.Layout(title='Óbitos por mês no nordeste: ',
                                 yaxis={'title': 'Óbitos novos'},
                                 xaxis={'title': 'Mês'})

#objeto figura

fig = go.Figure(data=nordeste, layout=configuracoes_layout)

#plotando o grafico
py.iplot(fig)
pyo.plot(fig)
```
