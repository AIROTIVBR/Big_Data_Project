```python
colunas_excluir =['codmun','codRegiaoSaude','nomeRegiaoSaude','casosAcumulado','casosNovos','Recuperadosnovos','emAcompanhamentoNovos','semanaEpi','Unnamed: 0','municipio','coduf','populacaoTCU2019']
df=df.drop(colunas_excluir, axis=1)
```

```python
df
```

```python
pernambuco = df[df.estado == 'PE']
pernambuco
```

```python
df.data = pd.to_datetime(df.data)
```

```python
df['data_mes'] = df.data.dt.to_period('M').astype(str)
```

```python
mortes_mes = df.groupby(by='data_mes').obitosNovos.sum()
```

```python
df = [go.Bar(x=mortes_mes.index,
               y=mortes_mes.values,
               marker= {'color': ' darkcyan'})]

#criando layout
configuracoes_layout = go.Layout(title='Óbitos por mês em Pernambuco: ',
                                 yaxis={'title': 'Óbitos novos'},
                                 xaxis={'title': 'Mês'})

#objeto figura

fig = go.Figure(data=df, layout=configuracoes_layout)

#plotando o grafico
py.iplot(fig)
pyo.plot(fig)
```
